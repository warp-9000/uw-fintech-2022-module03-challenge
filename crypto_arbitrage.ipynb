#### 1. For each of the three dates, measure the arbitrage spread between the two exchanges by subtracting the lower-priced exchange from the higher-priced one. Then use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame, where the spread is greater than zero.

*NOTE*: For illustration, only one of the three dates is shown in the starter code below.

```python
# Initializing date variables so it's easy to swap between them

early_date = '2018-01-01'
middle_date = '2018-02-04'
late_date = '2018-03-07'
# late_date = '2018-03-29' - one trade
```

```python
# For the date early in the dataset, measure the arbitrage spread between the two exchanges
# by subtracting the lower-priced exchange from the higher-priced one

arbitrage_spread_early = coinbase['Close'].loc[early_date].copy() - bitstamp['Close'].loc[early_date].copy()
arbitrage_spread_middle = coinbase['Close'].loc[middle_date].copy() - bitstamp['Close'].loc[middle_date].copy()
arbitrage_spread_late = coinbase['Close'].loc[late_date].copy() - bitstamp['Close'].loc[late_date].copy()


# Use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame

if arbitrage_spread_early.describe()['mean'] > 0:
    print('Arbitrage Spread Early:\n', arbitrage_spread_early.describe(),'\n')

if arbitrage_spread_middle.describe()['mean'] > 0:
    print('Arbitrage Spread Middle:\n', arbitrage_spread_middle.describe(),'\n')

if arbitrage_spread_late.describe()['mean'] > 0:
    print('Arbitrage Spread Late:\n', arbitrage_spread_late.describe(), '\n')

```

#### 2. For each of the three dates, calculate the spread returns. To do so, divide the instances that have a positive arbitrage spread (that is, a spread greater than zero) by the price of Bitcoin from the exchange you’re buying on (that is, the lower-priced exchange). Review the resulting DataFrame.

```python
# For the date early in the dataset, calculate the spread returns by dividing the instances when the arbitrage spread is positive (> 0) 
# by the price of Bitcoin from the exchange you are buying on (the lower-priced exchange).

spread_return_early = arbitrage_spread_early[arbitrage_spread_early>0] / bitstamp['Close'].loc[early_date]
spread_return_middle = arbitrage_spread_middle[arbitrage_spread_middle>0] / bitstamp['Close'].loc[middle_date]
spread_return_late = arbitrage_spread_late[arbitrage_spread_late>0] / bitstamp['Close'].loc[late_date]

# Review the spread return DataFrame

print(spread_return_early.describe(include='all'),'\n')
print(spread_return_middle.describe(include='all'),'\n')
print(spread_return_late.describe(include='all'),'\n')
```

#### 3. For each of the three dates, narrow down your trading opportunities even further. To do so, determine the number of times your trades with positive returns exceed the 1% minimum threshold that you need to cover your costs.

```python
# For the date early in the dataset, determine the number of times your trades with positive returns 
# exceed the 1% minimum threshold (.01) that you need to cover your costs

profitable_trades_early = spread_return_early[spread_return_early > .01]
print(f"Number of profitable trades in the early dataset: {profitable_trades_early.describe()['count']}")

profitable_trades_middle = spread_return_middle[spread_return_middle > .01]
print(f"Number of profitable trades in the middle dataset: {profitable_trades_middle.describe()['count']}")

profitable_trades_late = spread_return_late[spread_return_late > .01]
print(f"Number of profitable trades in the late dataset: {profitable_trades_late.describe()['count']}\n")


# Review the first five profitable trades
print(profitable_trades_early.head(5))
print(profitable_trades_middle.head(5))
print(profitable_trades_late.head(5))

```

#### 4. Generate the summary statistics of your spread returns that are greater than 1%. How do the average returns compare among the three dates?

```python
# For the date early in the dataset, generate the summary statistics for the profitable trades
# or you trades where the spread returns are are greater than 1%

print(profitable_trades_early.describe(include='all'),'\n')
print(profitable_trades_middle.describe(include='all'),'\n')
print(profitable_trades_late.describe(include='all'),'\n')
```

#### 5. For each of the three dates, calculate the potential profit, in dollars, per trade. To do so, multiply the spread returns that were greater than 1% by the cost of what was purchased. Make sure to drop any missing values from the resulting DataFrame.

```python
# For the date early in the dataset, calculate the potential profit per trade in dollars 
# Multiply the profitable trades by the cost of the Bitcoin that was purchased

profit_early = profitable_trades_early * bitstamp['Close'].loc[early_date]
profit_middle = profitable_trades_middle * bitstamp['Close'].loc[middle_date]
profit_late = profitable_trades_late * bitstamp['Close'].loc[late_date]

# Drop any missing values from the profit DataFrame

profit_per_trade_early = profit_early.dropna()
profit_per_trade_middle = profit_middle.dropna()
profit_per_trade_late = profit_late.dropna()

# View the early profit DataFrame

print('Profit Per Trade Early:\n',profit_per_trade_early,'\n')
print('Profit Per Trade Middle:\n',profit_per_trade_middle,'\n')
print('Profit Per Trade Late:\n',profit_per_trade_late,'\n')
```

```python
# What if we tried to arbitrage the whole range?
date_range = '2018-04-01'

# Calculating arbitrage spread between Coinbase and Bitstamp
arbitrage_spread_total = coinbase['Close'].loc[:date_range].copy() - bitstamp['Close'].loc[:date_range].copy()

# Extracting positive spreads
spread_return_total = arbitrage_spread_total[arbitrage_spread_total>0] / bitstamp['Close'].loc[:date_range]

# Finding profitable trades (those with > 1% spread)
profitable_trades_total = spread_return_total[spread_return_total > .01]

# Calculating the total profit based on price
profit_total = profitable_trades_total * bitstamp['Close'].loc[:date_range]

# Drop any missing values from the profit DataFrame
profit_per_trade_total = profit_total.dropna()

# Calculate the total profit over the whole range
profit_total = profit_per_trade_total.sum()
print(f"Profit Total: {profit_total}")

# Generate a DataFrame with the cumulative total profit
cumulative_profit_total = profit_per_trade_total.cumsum()

# Plotting the cumulative total profit
cumulative_total_ax = cumulative_profit_total.plot(
    legend=True, 
    title="Profit per Trade", 
    xlabel='Days',
    ylabel='Profit',
    kind='line',
    figsize=(12,7.5),
    grid=False,
    color='red',
    fontsize=12
)

cumulative_total_ax.legend(
    scatterpoints=1,
    loc='best',
    ncol=1,
    fontsize=14
)
```

**Some insight based on this graph...**

It's interesting that, based on the data given, most of the profit was generated in January shortly after the 15th. After that point there are only a few (relatively) small steps up in profit.
